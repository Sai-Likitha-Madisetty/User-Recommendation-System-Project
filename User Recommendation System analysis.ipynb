## Analyzing data with no issues

```python
import json

def load_data(filename):
    with open(filename, "r") as file:
        data = json.load(file)
    return data

def print_connections(main_data):
    print("\nPrinting Connections \n")
    for user in main_data["users"]:
        print(f"ID: {user['id']} - Name : {user['name']} - Friends : {user['friends']} - Liked pages : {user['liked_pages']}")
        
    print("\nPages: \n")
    for page in main_data["pages"]:
        print(f"{page['id']} - {page['name']}")

main_data = load_data("data_with_no_issues.json")
print_connections(main_data)
```

## Analyzing data with issues

```python
import json
def load_data():
    with open("data_with_issues.json", "r") as file:
        data = json.load(file)
    return data

def cleaning(python_data):
    # Remove empty named users
    python_data["users"] = [user for user in python_data["users"] if user["name"].strip()]
    
    # Remove duplicate friends
    for frnd in python_data["users"]:
        frnd["friends"] = sorted(list(set(frnd["friends"])))
        
    # Remove inactive users
    python_data["users"] = [user for user in python_data["users"] if user["friends"] or user["liked_pages"]]
    
    # Removing duplicate pages
    cleaned_dict = {}
    for page in python_data["pages"]:
        cleaned_dict[page["id"]] = page
    python_data["pages"] = list(cleaned_dict.values())

    return python_data

# Dumping data
def dumping_data():
    with open("Cleaned_data.json", "w") as file:
        final_data = json.dump(python_data, file)
    return final_data
    
python_data = load_data()
print(json.dumps(cleaning(python_data), indent=4))
```

## Friend Recommendations

```python
import json

def load_data():
    with open("cleaned_data", "r") as file:
        return json.load(file)

# MAIN LOGIC
def people_you_may_know(user_id, python_data):
    user_friends = {}
    for user in python_data["users"]:
        user_friends[user["id"]] = set(user["friends"]) # For id 1 - 2,3 are friends
    
    if user_id not in user_friends:
        return []
    
    direct_friends = user_friends[user_id] # 2,3
    score = {}
    for df in direct_friends:
        for mutual in user_friends.get(df, []):
            if mutual != user_id and mutual not in direct_friends:
                score[mutual] = score.get(mutual, 0) + 1 # {4,1}  after doing .items() we get [(4,1), (num1, num2), (5,9)]  like tuples inside a list. Here in 4,1 we have 1 right i.e., score of having chances to become friends with
                # According to that score we suggest with the order
    sorted_score = sorted(score.items(), key = lambda x: x[1], reverse = True)
    return [i for i, _ in sorted_score]
         

user_id = 10
python_data = load_data()
recommendations = people_you_may_know(user_id, python_data)
print(f"People you may know for user {user_id}: {recommendations}")
```

## Similar liked pages recommendation

```python
import json

def load_data():
    with open("cleaned_data", "r") as f:
        return json.load(f)
    
# MAIN LOGIC 
def pages_you_might_like(user_id, data):
    user_pages = {}
    for user in data["users"]:
        user_pages[user["id"]] = set(user["liked_pages"])
        
    user_liked_pages = user_pages[user_id]
    
    pages_score = {}
    
    for others, page in user_pages.items():
        if others != user_id:
            shared_pages = user_liked_pages.intersection(page) # 101.intersection(102) = {} = length of this will be 0
            for i in page:
                if i not in user_liked_pages:
                    pages_score[i] = pages_score.get(i, 0) + len(shared_pages)
    sorted_pages_score = sorted(pages_score.items(),key = lambda x: x[1], reverse = True)
    # return [j for j, _ in sorted_pages_score]
    return [(page_ID, score) for page_ID, score in sorted_pages_score]
                    

data = load_data()
user_id = 1
recommendations = pages_you_might_like(user_id, data)
print(f"Pages you might like for user with ID {user_id} is: {recommendations}")
```
